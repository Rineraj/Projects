Combining two subsequent rows for the same activity into one with start and complete timestamp

```python
event_number = 0
new_columns =  ['EventNumber', 'ActivityName', 'eLetter_ID', 'eLetter_Type', 'StartTime', 'EndTime']
new_event_log = pd.DataFrame(columns = new_columns)
activityname = None
eletterid = None

event_log.iloc[[0]]
for index, row in event_log.iterrows():
    if index == 0:
        activityname = row['ActivityName']
        eletterid = row['eLetter_ID']
        
        new_event_log = new_event_log.append({
            'EventNumber': event_number,
            'ActivityName' : row['ActivityName'], 
            'eLetter_ID' : row['eLetter_ID'],
            'eLetter_Type': row['eLetter_Type'],
            'StartTime': row['TotalTime']
        } , ignore_index=True)
        event_number = event_number + 1
    else:
        if (row['ActivityName']==activityname and row['eLetter_ID']==eletterid):
            new_event_log.loc[new_event_log.index[-1], 'EndTime']= row['TotalTime']
        else:
            activityname = row['ActivityName']
            eletterid = row['eLetter_ID']
        
            event_number = event_number + 1
            new_event_log = new_event_log.append({
                'EventNumber': event_number,
                'ActivityName' : row['ActivityName'], 
                'eLetter_ID' : row['eLetter_ID'],
                'eLetter_Type': row['eLetter_Type'],
                'StartTime': row['TotalTime']
            } , ignore_index=True)
```

```python
from twython import Twython  
import json

with open("C:/Users/rrine/Desktop/twitter_credentials.json", "r") as file:  
    creds = json.load(file)

# Instantiate an object
python_tweets = Twython(creds['CONSUMER_KEY'], creds['CONSUMER_SECRET'])


# Create our query
query = {'q': 'penguins',  
        'result_type': 'popular',
        'count': 100,
        'lang': 'en',
        'until': '2020-11-29'
        }


# Search tweets
dict_ = {'user': [], 'date': [], 'text': [], 'favorite_count': []}  
for status in python_tweets.search(**query)['statuses']:  
    dict_['user'].append(status['user']['screen_name'])
    dict_['date'].append(status['created_at'])
    dict_['text'].append(status['text'])
    dict_['favorite_count'].append(status['favorite_count'])

# Structure data in a pandas DataFrame for easier manipulation
df = pd.DataFrame(dict_)  
df.sort_values(by='favorite_count', inplace=True, ascending=False) 
df.head(5)

import pytz
df['changed_date'] = df.apply(lambda row: datetime.datetime.strptime(row['date'],'%a %b %d %H:%M:%S +0000 %Y').replace(tzinfo=pytz.UTC), axis=1)
df['latest_date'] = df.apply(lambda row: row['changed_date'].strftime('%Y-%m-%d'), axis=1)
twitter = df.groupby('latest_date').agg({'favorite_count': ['sum']}).reset_index()
twitter.columns = ['latest_date', 'favorite_count']
```

```python
penguin_count = new_event_log.copy()
penguin_count['latest_date'] = penguin_count.apply(lambda row: row['StartTime'].strftime('%Y-%m-%d'), axis=1)
```

```python
final_event_log = pd.merge(penguin_count, twitter, how='left', left_on=['latest_date'], right_on = ['latest_date'])
```
